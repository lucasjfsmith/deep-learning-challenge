## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# PARAMETER UPDATE 1: lower value counts thresshold from 500 to 25
app_type_mask = application_df['APPLICATION_TYPE'].value_counts() < 25

application_types_to_replace = list(application_df['APPLICATION_TYPE'].value_counts()[app_type_mask].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# PARAMETER UPDATE 2: lower value counts thresshold from 1000 to 100
classification_mask = application_df['CLASSIFICATION'].value_counts() < 100

classifications_to_replace = list(application_df['CLASSIFICATION'].value_counts()[classification_mask].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# PARAMETER UPDATE 3: investigate removing `STATUS` feature
print(application_df['STATUS'].value_counts())

# Because there are only 5 values of `0` this feature is likely just noise so we will remove it
application_df = application_df.drop(columns='STATUS', axis=1)
application_df.head()
```

```python
# PARAMETER UPDATE 4: investigate removing `SPECIAL_CONSIDERATIONS` feature
print(application_df['SPECIAL_CONSIDERATIONS'].value_counts())

# # Because there are only 27 values of `0` this feature is likely just noise so we will remove it
# application_df = application_df.drop(columns='SPECIAL_CONSIDERATIONS', axis=1)
# application_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
cat_cols = ['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','INCOME_AMT', 'SPECIAL_CONSIDERATIONS']
dummy_df = pd.get_dummies(application_df[cat_cols])
application_features_df = pd.concat([application_df['ASK_AMT'], dummy_df], axis=1)
application_features_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_features_df
y = application_df['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_dim = X_train_scaled.shape[1]
layer_units = 21

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=50, activation="relu", input_dim=input_dim))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=25, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=10, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity_Optimization.h5', save_format='h5')
```
